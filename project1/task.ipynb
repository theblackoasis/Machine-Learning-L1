```python
import feedparser
import pandas as pd
```

```python
d = feedparser.parse("http://www.moneycontrol.com/rss/marketreports.xml")
```

```python
posts = []
```

```python
for item in d.entries:
    a,description = item.description.split('/>')
    title = item.title.replace(";", ",")
    posts.append((item.published, title, item.link, description))
```

```python
df = pd.DataFrame(posts, columns=['timestamp', 'title', 'link', 'text'])
df.index.name="index"
```
